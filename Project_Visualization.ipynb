```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("Total.csv")
df.head()
```

```python
df.rename(columns = {'Partner':'country', 'Trade Value (US$)':"Trade_Value_(US$)", "Year": "year"}, inplace = True)
```

```python
#df_gapminder = px.data.gapminder()
df_gapminder = pd.read_csv("data.csv")
df_gapminder.head()
df_gapminder = df_gapminder.query('year >= 2003')
```

```python
df_gapminder.head()
```

```python
df_new = df_gapminder.merge(df, how='inner', on=["country", "year"])
```

```python
df_new.tail()
```

```python
df_new.to_csv("new.csv")
```

```python
df_export = df_new.query('Flow == "Export"')
```

```python
trade_min = df_export['Trade_Value_(US$)'].min()
trade_max = df_export['Trade_Value_(US$)'].max()

gdp_min =  df_export['gdpPercap'].min()
gdp_max =  df_export['gdpPercap'].max()
```

**Total Exports Year 2007**

```python
fig = px.scatter(df_export, x='gdpPercap', y='Trade_Value_(US$)', color='continent', size='pop', size_max=40, 
                hover_name='country', log_x=True, animation_frame='year',
                 animation_group='country', range_x=[gdp_min,gdp_max], range_y=[-5000,1000000000])
fig.update_layout(
    autosize=False,
    width=1000,
    height=400,
    margin=dict(
        l=50,
        r=50,
        b=100,
        t=100,
        pad=4
    ),
    paper_bgcolor="LightSteelBlue",
)
fig.write_html("Total_Exports_Year_2007.html")
fig.show()
```

```python
df_import = df_new.query('Flow == "Import"')
```

```python
df_import.tail()
```

```python
trade_min = df_import['Trade_Value_(US$)'].min()
trade_max = df_import['Trade_Value_(US$)'].max()

gdp_min =  df_import['gdpPercap'].min()
gdp_max =  df_import['gdpPercap'].max()
```

**Total Imports Year 2007**

```python
fig = px.scatter(df_import, x='gdpPercap', y='Trade_Value_(US$)', color='continent', size='pop', size_max=40, 
                hover_name='country', log_x=True, animation_frame='year',
                 animation_group='country', range_x=[gdp_min,gdp_max], range_y=[-5000,trade_max*1.25])
fig.update_layout(
    autosize=False,
    width=1000,
    height=400,
    margin=dict(
        l=50,
        r=50,
        b=100,
        t=100,
        pad=4
    ),
    paper_bgcolor="LightSteelBlue",
)
fig.write_html("Total_Imports_Year_2007.html")
fig.show()
```

```python
df_gapminder = pd.read_csv("data.csv")
```

**Comparison Between Life Expectancy, GDP and Population of countries.**

```python
# customize the labels 
fig = px.scatter(df_gapminder, x='gdpPercap', y='lifeExp', color='continent', size='pop', size_max=40, 
                hover_name='country',log_x=True, animation_frame='year',
                 animation_group='country', range_x=[25, 10000], range_y=[25,90], 
                labels=dict(pop="Population", gdpPercap="GDP Per Capita", lifeExp="Life Expectency"))
fig.update_layout(
    autosize=False,
    width=1000,
    height=400,
    margin=dict(
        l=50,
        r=50,
        b=100,
        t=100,
        pad=4
    ),
    paper_bgcolor="LightSteelBlue",
)
fig.write_html("lifexp_gdp__pop.html")
fig.show()
```

**GDP, Country Comparison Per Year**

```python
# create a map using choropleth
fig = px.choropleth(df_gapminder, locations='iso_alpha', color='gdpPercap', hover_name='country', 
                    animation_frame='year', color_continuous_scale=px.colors.sequential.Plasma, projection='natural earth')
fig.write_html("gdp_country_year.html")
fig.show() 
```

```python
df_export_Pakistan = pd.read_csv("export.csv")
df_import_Pakistan = pd.read_csv("import.csv")
```

```python
df_export_Pakistan["Flow"] = "Export"
df_import_Pakistan["Flow"] = "Import" 
df_export_Pakistan = df_export_Pakistan[(df_export_Pakistan['Partner'] != "World") & (df_export_Pakistan['Trade Value (US$)'] > 0)]
df_import_Pakistan = df_import_Pakistan[(df_import_Pakistan['Partner'] != "World") & (df_import_Pakistan['Trade Value (US$)'] > 0)]
df_total = df_import_Pakistan.append(df_export_Pakistan)
```

```python
df = df_total[["Year", "Trade Value (US$)", "Partner", "Flow"]]
df.to_csv("Total.csv")
```

```python
df = df_export_Pakistan[["Year", "Trade Value (US$)", "Partner"]]
```

```python
df_filtered = df[(df['Partner'] != "World") & (df['Trade Value (US$)'] > 0)]
df_filtered.head()
```

```python
df_filtered.columns
```

```python
minimum = df_filtered["Trade Value (US$)"].min()
maximum = df_filtered["Trade Value (US$)"].max()
```

```python
import plotly.express as px

lis = px.colors.sequential.Rainbow
print(lis)
```

```python
# create a map using choropleth
fig = px.choropleth(df_filtered, locations='Partner', locationmode='country names', color='Trade Value (US$)', hover_name='Partner', 
                    animation_frame='Year', projection='natural earth',
                    range_color = [minimum, maximum],
                    color_continuous_scale = lis)
fig.write_html("file2.html")
fig.show() 
```
